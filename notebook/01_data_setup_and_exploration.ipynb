```python
def uniqueness(df):
    for i in df.columns[1:]:
        if len(df[i].unique()) <= 8:
            print(f'Column ~{i} ~> {df[i].unique()}')
```

```python

# Convert 'TotalCharges' to numeric, coercing errors (e.g., blank spaces)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Replace NaN (caused by blank/invalid values) with the mean
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)
```

```python
df['TotalCharges']=df['TotalCharges'].astype(float)
```

# Data cleaning and Preprocessing day-02

```python
def find_null(df):
    for i in df.columns[1:]:

        print(f'Column ~{i} ~> {df[i].isna().sum()}')
```

```python
find_null(df)
```

```python

cols = [ "Partner", "Dependents", "PhoneService","PaperlessBilling","Churn" ]

for col in cols:
    df[col] = df[col].replace({"Yes": 1, "No": 0})
```

```python
df
```

```python
df=df.drop("customerID", axis=1)
```

```python

categorical_cols = ['gender',
    'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
    'Contract', 'PaymentMethod'
]
```

```python


# Assuming df is your original DataFrame
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python

bool_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[bool_cols] = df_encoded[bool_cols].astype(int)
```

```python

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# Define preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['MonthlyCharges', 'TotalCharges'])
    ],
    remainder='passthrough'  # Keep all other columns as-is
)
```

```python
# Apply transformation
X_scaled = preprocessor.fit_transform(df_encoded)

# Retrieve correct feature names
feature_names = preprocessor.get_feature_names_out()
X_scaled_df = pd.DataFrame(X_scaled, columns=feature_names)

# Optional: Clean up column names
X_scaled_df.columns = [col.split("__")[-1] for col in X_scaled_df.columns]

X_scaled_df.head()
```

```python
X_scaled_df.to_csv("telco_transformed.csv", index=False)
```
